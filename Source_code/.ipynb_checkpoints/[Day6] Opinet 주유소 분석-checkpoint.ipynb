```python
import pandas as pd
```

```python
##glob - 파일 경로 접근 시 사용
from glob import glob
```

```python
stations_files=glob('../data/05/지역*.xls')

tmp_raw=[]

# for 문을 사용하여 엑셀 데이터를 불러서 tmp_raw list에 넣기
for file in stations_files:
#     print(file)
    tmp = pd.read_excel(file, header=2)
    tmp_raw.append(tmp)
    
tmp_raw
    
```

```python
station_raw=pd.concat(tmp_raw)
#concat으로 붙여서 데이터프레임으로 만드는거
station_raw
```

```python
stations = pd.DataFrame({'Oil_store':station_raw['상호'],
                         '주소':station_raw['주소'],
                         '휘발유':station_raw['휘발유'],
                         '셀프여부':station_raw['셀프여부'],
                         '상표':station_raw['상표']})
stations.head()
```

```python
stations.head().info()
```

```python
## 주소에서 구만 빼서 column을 새로 만들어서 저장

stations['구']=[address.split()[1] for address in stations['주소']
```
