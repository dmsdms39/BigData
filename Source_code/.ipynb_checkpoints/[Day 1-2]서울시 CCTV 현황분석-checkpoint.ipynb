### numpy (cf) 피어슨 상관계수 (corrcoef)

두개를 x축, y축으로 두고, 점의 연결선(추이선)을 그렸을때 기울기 = 상관관계, 0이나 1에 가까울 때 상관관계
- 절대값 0.1 이하면 무시
- 절대값 0.3 이하면 약한 상관관계
- 절대값 0.7 이하면 뚜렷한 상관관계

```python
import numpy as np
np.corrcoef(data_result['고령자 비율'],data_result['소계'])
#[자기자신(0,0), 서로(0,1)]
#[서로(1,0), 자기자신(1,1)]
#(0,0)(1,1)은 볼 필요 없음
```

```python
np.corrcoef(data_result['외국인 비율'],data_result['소계'])
```

```python
np.corrcoef(data_result['인구수'],data_result['소계'])
```

```python
plt.figure(figsize=(5,5))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV 수')
plt.grid()
plt.show()

##추세선을 그려야 그래프의 의미가 생김
```

```python
fp1= np.polyfit(data_result['인구수'],data_result['소계'],1)#맨마지막숫자 높이면 기본항이 많아짐->2차선, 3차선
print(fp1)
##polyfit은 추세선의 계수
```

```python
f1= np.poly1d(fp1)
print(f1)
fx = np.linspace(100000,700000,100) #linspace(간격, 선의 길이, 몇개 생성) 10000~70000까지 100개 만들기
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=40)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV 수')
plt.grid()
plt.show()
```

## Poly1D : one-dimensional polynomial class

### 예시
= `(x+1) + (x-1)` <- `np.poly1d([1,1]) + np.poly1d([1,-1])`
= `(x+1) * (x-1)` <- `np.poly1d([1,1]) * np.poly1d([1,-1])`

```python
#오차 계산하기
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
##수치를 abs 절대값으로 만들어줌
df_sort = data_result.sort_values(by='오차', ascending = False)
df_sort.head()
```

```python
plt.figure(figsize=(14,7))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)
# c=오차값이므로 멀어질수록 숫자큰 노란색
plt.plot(fx,f1(fx),ls='dashed',lw=4)

##range(len(df_sort))로 하면 전체 볼수 있지만
##오차 큰 순서대로 sorting을 미리 해뒀기 때문에 아래 그래프와 같이 나옴
for n in range(10):
    plt.text(df_sort['인구수'][n]*1.01, df_sort['소계'][n]*0.99, df_sort.index[n], fontsize=12) ##plt.text(x,y,string,fontsize) *위치설정

plt.xlabel('인구수')
plt.ylabel('인구당 cctv 비율')

plt.colorbar()
plt.grid()
plt.show()
```

```python
##테스트
fp2= np.polyfit(data_result['고령자 비율'],data_result['소계'],3)
print(fp2)
f2= np.poly1d(fp2)
print(f2)
fx = np.linspace(12,20,100)
plt.figure(figsize=(20,10))
plt.scatter(data_result['고령자 비율'], data_result['소계'], s=40)
plt.plot(fx,f2(fx), ls='dashed', lw=3, color='g')

df1_sort = data_result.sort_values(by='고령자 비율', ascending = False)

for n in range(len(df1_sort)):
    plt.text(df1_sort['고령자 비율'][n]*1.003, df1_sort['소계'][n]*0.975, df1_sort.index[n], fontsize=12)
    
plt.xlabel('테스트 x축')
plt.ylabel('테스트 y축')
plt.grid()
plt.show()
```

```python
##결과 저장하기
df_sort.to_csv('../data/01.cctv_result.csv', sep=',', encoding='utf-8')
df1_sort.to_csv('../data/01.test.csv', sep=',', encoding='utf-8')
```
