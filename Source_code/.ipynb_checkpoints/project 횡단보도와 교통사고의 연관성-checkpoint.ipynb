```python
import pandas as pd

street_lamp = pd.read_excel('../data/sp_street_lamp.xls', header=9, usecols='B, AD', encoding = 'utf-8')
street_lamp.rename(columns={street_lamp.columns[0]:'구별', street_lamp.columns[1]:'가로등'},inplace=True)
street_lamp
```

```python
cross_walk = pd.read_excel('../data/sp_cross_walk.xlsx', header=1, usecols='B, C, D, E', encoding = 'utf-8')
cross_walk.drop([0], inplace=True)
cross_walk.rename(columns={cross_walk.columns[0]:'구별', 
                           cross_walk.columns[1]:'횡단보도',
                          },inplace=True)
cross_walk
```

```python
traffic_accident = pd.read_excel('../data/sp_traffic_accident.xls', header=1, usecols='B, D, E, C', encoding = 'utf-8')
traffic_accident.rename(columns={traffic_accident.columns[0]:'구별', traffic_accident.columns[1]:'발생건수',traffic_accident.columns[2]:'발생건수(만대당)', traffic_accident.columns[3]:'사망자수'},inplace=True)
traffic_accident
```

```python
traffic_accident['사건당 사망자수']=traffic_accident['사망자수'] / traffic_accident['발생건수']
traffic_accident
```

```python
car_speed = pd.read_excel('../data/sp_car_speed.xls', header=3, usecols='C, D', encoding = 'utf-8')
car_speed.rename(columns={car_speed.columns[0]:'구별', car_speed.columns[1]:'주행속도(km/h)'},inplace=True)
car_speed
```

```python
data_result = pd.merge(cross_walk ,traffic_accident, on='구별')
data_result
```

```python
data2_result = pd.merge(data_result ,car_speed, on='구별')
data2_result
```
